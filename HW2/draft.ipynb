```python
import numpy as np
# 1 & 3 & -1 & 0 \\
# 2 & 1 & 0 & -1 \\
# 1 & 2 & 0 & 0 \\
# 3 & 1 & 0 & 0 \\
B = [[1, 3, -1, 0],
     [2, 1, 0, -1],
     [1, 2, 0, 0],
     [3, 1, 0, 0]]
b = [15, 10, 40, 60]

# 0 & 0 \\
# 0 & 0 \\ 
# 1 & 0 \\
# 0 & 1 \\
N = [[0, 0],
     [0, 0],
     [1, 0],
     [0, 1]]

np.linalg.solve(B, b)
```

```python
cN = [0, 0]
cB = [-1, -1, 0, 0]

cN - cB @ np.linalg.inv(B) @ N
```

```python
dN = [1, 0]
dB = - np.linalg.inv(B) @ N @ dN

xB = np.linalg.solve(B, b)

theta_list = - xB / dB
theta_list
```
